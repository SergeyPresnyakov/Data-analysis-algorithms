```python
plt.style.use('fivethirtyeight')

n = X.shape[1]
alpha = 1e-3
w = np.array([1, 0.5])
plt.figure(figsize=(10, 7))
for i in range(10001):
    y_pred = np.dot(w, X)
    err = calc_mse(y, y_pred)
    for j in range(w.shape[0]):
        w[j] -= alpha * (1/n * 2 * np.sum(X[j] * (y_pred - y)))
    if i % 1000 == 0:
#         print(w.shape[0])
        print(f"Номер итерации: {i}, веса модели: {w}, MSE: {err}")
        plt.scatter(X[1], y)
        plt.plot(X[1], w[0] + w[1] * X[1], color = cnames[int(i / 100)+2], label=i)
        plt.legend()
```
